
### Q6. Analyze the dataset to identify the peak travel season. What features would you analyze, and how would you present your findings?

To identify the peak travel season, we would analyze the date-related features such as the departure date and the number of bookings per month. Heres a step-by-step approach:

1. Extract month from the departure date.
2. Group by month and count the number of bookings.

```python
# Extract month from the departure date
df['Month'] = pd.to_datetime(df['Date_of_Journey']).dt.month

# Group by month and count the number of bookings
monthly_bookings = df['Month'].value_counts().sort_index()

# Plot the number of bookings per month
plt.figure(figsize=(10, 6))
monthly_bookings.plot(kind='bar')
plt.title('Number of Bookings per Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.grid(True)
plt.show()
```

### Q7. Identify trends in flight prices. What features would you analyze, and what visualizations would you use?

To identify trends in flight prices, we would analyze features like date of booking, day of the week, and seasonality. We can visualize these trends using line plots and bar charts.

1. Extract the day of the week and month from the departure date.
2. Calculate the average price per day of the week and per month.


```python
# Extract day of the week from the departure date
df['Day_of_Week'] = pd.to_datetime(df['Date_of_Journey']).dt.dayofweek

# Calculate the average price per day of the week
average_price_day = df.groupby('Day_of_Week')['Price'].mean()

# Plot the average price per day of the week
plt.figure(figsize=(10, 6))
average_price_day.plot(kind='line')
plt.title('Average Flight Price by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()

# Calculate the average price per month
average_price_month = df.groupby('Month')['Price'].mean()

# Plot the average price per month
plt.figure(figsize=(10, 6))
average_price_month.plot(kind='line')
plt.title('Average Flight Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

### Q8. Identify factors that affect flight prices. What features would you analyze, and how would you present your findings?

To identify factors that affect flight prices, we would analyze features such as:

- Airline
- Source and destination cities
- Duration of the flight
- Total stops
- Class of travel (if available)

Here's a basic analysis approach:

1. Calculate the average price for different airlines, cities, and number of stops.
2. Use boxplots and bar charts to visualize the impact of these features on flight prices.


```python




# Calculate the average price for different airlines
average_price_airline = df.groupby('Airline')['Price'].mean()

# Plot the average price for different airlines
plt.figure(figsize=(12, 8))
average_price_airline.plot(kind='bar')
plt.title('Average Flight Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# Calculate the average price for different source cities
average_price_source = df.groupby('Source')['Price'].mean()

# Plot the average price for different source cities
plt.figure(figsize=(10, 6))
average_price_source.plot(kind='bar')
plt.title('Average Flight Price by Source City')
plt.xlabel('Source City')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()

# Calculate the average price for different destination cities
average_price_destination = df.groupby('Destination')['Price'].mean()

# Plot the average price for different destination cities
plt.figure(figsize=(10, 6))
average_price_destination.plot(kind='bar')
plt.title('Average Flight Price by Destination City')
plt.xlabel('Destination City')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()

# Calculate the average price for different number of stops
average_price_stops = df.groupby('Total_Stops')['Price'].mean()

# Plot the average price for different number of stops
plt.figure(figsize=(10, 6))
average_price_stops.plot(kind='bar')
plt.title('Average Flight Price by Number of Stops')
plt.xlabel('Number of Stops')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()

```


### Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

Heres the code for a bar chart:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
google_playstore_df = pd.read_csv(url)

# Replace non-numeric 'Price' values with NaN, then convert to float
google_playstore_df['Price'] = google_playstore_df['Price'].str.replace('$', '')
google_playstore_df['Price'] = pd.to_numeric(google_playstore_df['Price'], errors='coerce')

# Calculate average price by app type, excluding NaN values
average_price_by_type = google_playstore_df.groupby('Type')['Price'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
average_price_by_type.plot(kind='bar')
plt.title('Average App Price by Type')
plt.xlabel('Type')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```


### Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.


```python
import pandas as pd

# Load the dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
google_playstore_df = pd.read_csv(url)

# Replace non-numeric 'Installs' values with NaN, then convert to int
google_playstore_df['Installs'] = google_playstore_df['Installs'].str.replace(',', '').str.replace('+', '')
google_playstore_df['Installs'] = pd.to_numeric(google_playstore_df['Installs'], errors='coerce')

# Drop rows with NaN values in 'Installs' column
google_playstore_df = google_playstore_df.dropna(subset=['Installs'])

# Convert 'Installs' to int
google_playstore_df['Installs'] = google_playstore_df['Installs'].astype(int)

# Get the top 10 most popular apps
top_10_apps = google_playstore_df[['App', 'Installs']].sort_values(by='Installs', ascending=False).head(10)
top_10_apps
```

### Q15. Identify the most popular app categories to make recommendations to the company.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = 'https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv'
google_playstore_df = pd.read_csv(url)

# Replace non-numeric 'Installs' values with NaN, then convert to int
google_playstore_df['Installs'] = google_playstore_df['Installs'].str.replace(',', '').str.replace('+', '')
google_playstore_df['Installs'] = pd.to_numeric(google_playstore_df['Installs'], errors='coerce')

# Drop rows with NaN values in 'Installs' column
google_playstore_df = google_playstore_df.dropna(subset=['Installs'])

# Convert 'Installs' to int
google_playstore_df['Installs'] = google_playstore_df['Installs'].astype(int)

# Calculate the total number of installs by category
installs_by_category = google_playstore_df.groupby('Category')['Installs'].sum().sort_values(ascending=False)

# Plot the total number of installs by category
plt.figure(figsize=(14, 8))
installs_by_category.plot(kind='bar')
plt.title('Total Number of Installs by App Category')
plt.xlabel('Category')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Q17. Identify the best time to launch a new app.
To identify the best time to launch a new app, we could analyze the number of app releases over time:


```python
# Extract the month from the 'Last Updated' column
google_playstore_df['Last Updated'] = pd.to_datetime(google_playstore_df['Last Updated'], errors='coerce')


google_playstore_df['Month'] = google_playstore_df['Last Updated'].dt.month

# Calculate the number of app updates by month
updates_by_month = google_playstore_df['Month'].value_counts().sort_index()

# Plot the number of app updates by month
plt.figure(figsize=(10, 6))
updates_by_month.plot(kind='bar')
plt.title('Number of App Updates by Month')
plt.xlabel('Month')
plt.ylabel('Number of Updates')
plt.grid(True)
plt.show()
```

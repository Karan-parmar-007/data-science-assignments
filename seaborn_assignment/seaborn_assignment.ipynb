### **Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

Seaborn is a powerful visualization library in Python, built on top of Matplotlib, which provides a high-level interface for drawing attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn, along with their uses:

### 1. **Scatter Plot (`scatterplot`)**
**Uses:**
- **Exploring relationships between two continuous variables:** Scatter plots help visualize how one variable affects another.
- **Identifying trends, patterns, and correlations:** By plotting data points, you can see if there is a linear or non-linear relationship, clusters, or outliers.
- **Highlighting groups:** Different colors and markers can be used to represent different categories within the data.

### 2. **Bar Plot (`barplot`)**
**Uses:**
- **Comparing categorical data:** Bar plots are used to show the distribution of categorical data.
- **Aggregating values:** They can display aggregated values such as means, sums, or counts for different categories.
- **Visualizing counts and proportions:** Useful for showing the count or proportion of items in different categories.

### 3. **Histogram (`histplot`)**
**Uses:**
- **Visualizing the distribution of a single variable:** Histograms show the frequency distribution of continuous data.
- **Identifying the shape of the data distribution:** They help to see if the data is normally distributed, skewed, or has outliers.
- **Comparing distributions:** By overlaying multiple histograms, you can compare the distributions of different datasets.

### 4. **Box Plot (`boxplot`)**
**Uses:**
- **Summarizing the distribution of a variable:** Box plots display the median, quartiles, and potential outliers of a dataset.
- **Comparing across categories:** They allow for comparison of the distribution of data across multiple categories.
- **Identifying outliers:** Box plots highlight outliers, which are data points that fall outside the whiskers of the plot.

### 5. **Heatmap (`heatmap`)**
**Uses:**
- **Visualizing matrices:** Heatmaps are ideal for displaying data in matrix form, with colors representing the magnitude of values.
- **Correlation matrices:** They are commonly used to show correlation matrices, indicating how variables are related.
- **Pattern recognition:** Heatmaps can reveal patterns, correlations, and areas of high or low activity within a dataset.

These plots provide a wide range of visualization capabilities for different types of data, making Seaborn a versatile tool for data analysis and presentation.

### **Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

```python
import seaborn as sns
import matplotlib.pyplot as plt


fmri = sns.load_dataset("fmri")


plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event and Region")
plt.show()
```

### **Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.**

```python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")


plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')


plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')

plt.tight_layout()
plt.show()
```

### **Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Plot a histogram for the 'price' column with hue for the 'cut' column
plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', palette='muted')
plt.title('Price Distribution of Diamonds by Cut')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

### **Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Plot a pair plot with hue for the species column
sns.pairplot(iris, hue='species', palette='muted')
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02)
plt.show()
```

### **Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to get a matrix form for the heatmap
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Monthly Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

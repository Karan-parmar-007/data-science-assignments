Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

p = figure(title="Simple Bokeh Plot", x_axis_label='X-axis', y_axis_label='Y-axis')

p.circle(x, y, size=10, color='navy', alpha=0.5)

p.title.text_color = "green"
p.title.text_font = "times"
p.title.text_font_style = "italic"

output_file("bokeh_plot.html")
show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


In Bokeh, glyphs are the visual markers that represent data points on a plot. These glyphs can be various shapes such as circles, squares, triangles, lines, patches, etc. Glyphs are added to a Bokeh plot using methods provided by the figure object.

```python
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

p = figure(title="Glyphs Example", x_axis_label='X-axis', y_axis_label='Y-axis')

p.circle(x, y, size=10, color='navy', alpha=0.5)
p.line(x, y, line_width=2, color="red", alpha=0.6)

output_file("bokeh_glyphs_plot.html")
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

You can customize the appearance of a Bokeh plot, including the axes, title, and legend, by modifying various properties of the figure object. Here's how you can do it:

Title: You can set the title of the plot using the title property of the figure object.

Axes: You can customize the appearance of the axes by setting properties such as x_axis_label, y_axis_label, x_axis_type, y_axis_type, x_range, y_range, x_axis_location, y_axis_location, x_minor_ticks, y_minor_ticks, etc.

Legend: You can add a legend to the plot using the legend property of the figure object. You can customize the legend's position, orientation, background color, border color, text color, font size, etc.

```python
from bokeh.plotting import figure, output_file, show

# Prepare Data
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# Create a Figure
p = figure(title="Customized Bokeh Plot",
           x_axis_label='X-axis',
           y_axis_label='Y-axis',
           x_range=(0, 6),
           y_range=(5, 11),
           x_axis_location="above",
           y_axis_location="right",
           toolbar_location="below",
           tools="pan,wheel_zoom,box_zoom,reset",
           width=600,
           height=400)

# Add Glyphs
p.circle(x, y, size=10, color='navy', alpha=0.5, legend_label="Circle Glyph")
p.line(x, y, line_width=2, color="red", alpha=0.6, legend_label="Line Glyph")

# Customize Title
p.title.text_color = "green"
p.title.text_font = "times"
p.title.text_font_style = "italic"

# Customize Axes
p.xaxis.axis_label_text_color = "blue"
p.yaxis.axis_label_text_font_size = "12pt"

# Customize Legend
p.legend.location = "top_left"
p.legend.orientation = "horizontal"
p.legend.label_text_color = "black"
p.legend.label_text_font_size = "10pt"
p.legend.background_fill_color = "lightgray"
p.legend.border_line_color = "black"

# Show or Save the Plot
output_file("customized_bokeh_plot.html")
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a Python process that allows you to create, deploy, and manage interactive Bokeh plots and applications. It enables you to build dynamic, web-based visualizations that can be updated in real time based on user interactions, data changes, or other events.

With a Bokeh server, you can create complex interactive applications that go beyond what is possible with static HTML files alone. You can handle user input, streaming data, and complex interactions between different components of your visualization.

Here's how you can use a Bokeh server to create interactive plots that can be updated in real time:

Define the Application: Create a Python script that defines your Bokeh application. This script will typically include the layout and logic of your visualization, including any interactions and updates.

Create the Bokeh Server Application: Use the bokeh.application module to define your Bokeh server application. This typically involves creating a Document object that represents the content of your application.

Run the Bokeh Server: Use the bokeh.server.server.Server class to run the Bokeh server and host your application. You can start the server from the command line or within your Python script.

Access the Application: Once the Bokeh server is running, users can access your interactive visualization by navigating to the appropriate URL in their web browser. They can interact with the visualization in real time, and any updates or changes will be reflected immediately.

### **Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

### What is Matplotlib?

Matplotlib is a widely used open-source plotting library for the Python programming language. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK. Matplotlib is particularly well-suited for creating static, animated, and interactive visualizations in Python.

### Why is it used?

Matplotlib is used for a variety of reasons, including:
1. **Visualization**: It helps in visualizing data in a graphical format, making it easier to understand trends, patterns, and outliers.
2. **Versatility**: It supports a wide range of plots and chart types, from basic line and bar charts to complex 3D plots and custom visualizations.
3. **Integration**: It integrates well with other Python libraries such as NumPy, Pandas, and SciPy, making it a powerful tool for data analysis.
4. **Customization**: It offers extensive customization options for plots, allowing users to tweak almost every aspect of a figure.
5. **Community and Documentation**: It has a large user community and extensive documentation, making it easier for new users to get started and find solutions to problems.

### Five Plots Using Pyplot Module of Matplotlib

The `pyplot` module of Matplotlib provides a MATLAB-like interface for creating a wide variety of plots. Here are five types of plots that can be created using `pyplot`:

1. **Line Plot**:
   - **Purpose**: To display information as a series of data points called 'markers' connected by straight line segments.
   - **Example**:
     ```python
     import matplotlib.pyplot as plt
     import numpy as np

     x = np.linspace(0, 10, 100)
     y = np.sin(x)
     plt.plot(x, y)
     plt.xlabel('x axis')
     plt.ylabel('y axis')
     plt.title('Line Plot')
     plt.show()
     ```

2. **Bar Plot**:
   - **Purpose**: To represent categorical data with rectangular bars with lengths proportional to the values that they represent.
   - **Example**:
     ```python
     categories = ['A', 'B', 'C', 'D']
     values = [4, 7, 1, 8]
     plt.bar(categories, values)
     plt.xlabel('Category')
     plt.ylabel('Values')
     plt.title('Bar Plot')
     plt.show()
     ```

3. **Histogram**:
   - **Purpose**: To represent the distribution of a dataset by showing the number of data points that fall within a specified range of values (bins).
   - **Example**:
     ```python
     data = np.random.randn(1000)
     plt.hist(data, bins=30)
     plt.xlabel('Value')
     plt.ylabel('Frequency')
     plt.title('Histogram')
     plt.show()
     ```

4. **Scatter Plot**:
   - **Purpose**: To display values for typically two variables for a set of data, using Cartesian coordinates to display values for two variables.
   - **Example**:
     ```python
     x = np.random.rand(50)
     y = np.random.rand(50)
     plt.scatter(x, y)
     plt.xlabel('x axis')
     plt.ylabel('y axis')
     plt.title('Scatter Plot')
     plt.show()
     ```

5. **Pie Chart**:
   - **Purpose**: To represent data as a circular graph, which is divided into slices to illustrate numerical proportion.
   - **Example**:
     ```python
     sizes = [15, 30, 45, 10]
     labels = ['A', 'B', 'C', 'D']
     plt.pie(sizes, labels=labels, autopct='%1.1f%%')
     plt.title('Pie Chart')
     plt.show()
     ```

These are just a few examples of the many types of plots that can be created using Matplotlib's `pyplot` module. Each type of plot is useful for visualizing different kinds of data and can be customized extensively to meet specific needs.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
 
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

````python
import numpy as np
import matplotlib.pyplot as plt 
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
# colors = np.random.choice(['red', 'blue'], size=50)
# plt.scatter(x, y, c=colors)
x_color = 'red'
y_color = 'green'

# Create scatter plot
plt.scatter(x, y, c=[x_color if i % 2 == 0 else y_color for i in range(len(x))])
plt.xlabel('x axis')

plt.ylabel('y axis')

plt.title('Scatter Plot')

plt.show()
````

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function in Matplotlib is used to create multiple plots within a single figure. It allows for the creation of a grid of plots, where each plot can be accessed and modified independently. This is useful for comparing multiple datasets or visualizations side by side, making it easier to analyze and interpret complex data.



````python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])



fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1, 'r')
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x2, y2, 'g')
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')

axs[1, 0].plot(x3, y3, 'b')
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')

axs[1, 1].plot(x4, y4, 'm')
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')

plt.tight_layout()

plt.show()
````

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

**What is a Box Plot?**
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of a dataset: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The box represents the interquartile range (IQR), which is the range between the first and third quartiles. Whiskers extend from the box to the minimum and maximum values within a certain range, typically 1.5 times the IQR.

**Why is it Used?**
Box plots are used for several purposes:

Visualizing Distribution: Box plots provide a concise summary of the distribution of a dataset, including the central tendency, spread, and presence of outliers.
Comparison: They allow for easy comparison of distributions between different groups or datasets.
Identifying Outliers: They highlight potential outliers in a dataset, making it easier to identify extreme values.
Understanding Variability: They provide insights into the spread and variability of data points, particularly in comparison to the median and quartiles.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a single list
data = [box1, box2]

# Create a box plot
plt.boxplot(data)

# Add labels and title
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')

# Show plot
plt.show()
````
